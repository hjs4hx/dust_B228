# Search ALMA Archive: IRAS 15398

The purpose of this code is to search the ALMA archive for all observations for a particular set of coordinates.

You will need to most up to date version of astroquery (pip install -U --pre astroquery)

```python
#necessary imports

import astroquery
from astroquery.alma import Alma
from astropy import coordinates
from astropy import units as u
```

```python
# define the coordinates of your source:
IRAS15 = coordinates.SkyCoord('15:43:02.160 -34:09:09.000', frame='fk5', unit=(u.hour, u.deg))

#define the resulting table as 'result'. 
#parameter 'radius' is radius in arcsec for which you want the archive to return observations
result = Alma.query_region(IRAS15, radius=10.*u.arcsec)


#look at new coordinates from 2MASS source:
#NOTE: these coordinates are from simbad based on a search radius around non-exact coordinates from paper
SouthernIRAS15 = coordinates.SkyCoord('15:43:01.31640 -34:09:15.3036', frame='fk5', unit=(u.hour, u.deg))

result_southern = Alma.query_region(SouthernIRAS15, radius=10.*u.arcsec)
```
